# Basic Setup

```python
%matplotlib inline
%config InlineBackend.figure_format = 'retina'
import pandas as pd
import os
import matplotlib.pylab as plt
import numpy as np
import seaborn as sns
from natsort import natsorted, ns
import glob
from pymer4.models import Lmer
from sklearn import metrics
from sklearn.metrics import roc_auc_score
%load_ext rpy2.ipython

fpath = 'where/you/have/the/data'
```

```python
%%R
library('lme4')
library('lmerTest')
library('lsmeans')
```

# Patient - Interaction - Overall pain rating

```python
pat_data = pd.read_csv(os.path.join(fpath,'study1_pat_interaction_overall_pain_data.csv'))
pat_mean = pat_data.groupby(['DyadID','Condition','block'])['Pain_Rating'].mean().reset_index()
```

```python
pat_data['trials_c'] = pat_data.N - pat_data.N.mean()
```

```python
%%R -i pat_data
pat_data$Site <- factor(pat_data$Site)
pat_data$Condition <- factor(pat_data$Condition_new)
pat_data$Color <- factor(pat_data$Current_Color)
pat_overall_m <- lmer(Pain_Rating ~ Site + trials_c + Color + Condition
                  + (1 + trials_c + Condition | DyadID), data = pat_data)
summary(pat_overall_m)
```

## Figure 1(B-3)

```python
sns.set(style="ticks", font_scale=1.3)
g = sns.catplot(x="Condition", y="Pain_Rating", palette={'Thermedol':'dodgerblue','Vaseline':'red'},units='DyadID',
                   n_boot=5000, ci=68, data=pat_mean, kind="bar", order=['Vaseline','Thermedol'], aspect=.8)

g = sns.stripplot(x="Condition", y="Pain_Rating", palette={'Thermedol':'lightskyblue','Vaseline':'lightsalmon'},
                   data=pat_mean, order=['Vaseline','Thermedol'],alpha=.5)

g.set(ylim=(0, 70))
plt.ylabel('Pain rating')
plt.xlabel('')
g.set_xticklabels(["Control","Thermedol"])
```

# Patient - Interaction - Maximal pain from continuous pain rating

```python
cont_data = pd.read_csv(os.path.join(fpath,'study1_pat_interaction_maxcont_pain_data.csv'))
```

```python
cont_mean = cont_data.groupby(['DyadID','Condition','block'])['max_pain'].mean().reset_index()
```

```python
%%R -i cont_data
cont_data$Condition <- factor(cont_data$Condition)
cont_data$Color <- factor(cont_data$Current_Color)
cont_data$Site <- factor(cont_data$Site)
pat_cont_max_model <- lmer(max_pain ~ Site + Color + trials_c + Condition + (1 + trials_c + Condition | DyadID)
                           , data = cont_data)
summary(pat_cont_max_model)
```

## Extended Data Figure 2(B)

```python
sns.set(style="ticks", font_scale=1.3)
g = sns.catplot(x="Condition", y="max_pain", palette={'Thermedol':'dodgerblue','Vaseline':'red'},units='DyadID',
                   n_boot=5000, ci=68, data=cont_mean, kind="bar", order=['Vaseline','Thermedol'], aspect=.8)

g = sns.stripplot(x="Condition", y="max_pain", palette={'Thermedol':'lightskyblue','Vaseline':'lightsalmon'},
                   data=cont_mean, order=['Vaseline','Thermedol'],alpha=.5)

g.set(ylim=(0, 80))
plt.ylabel('Maximal Pain rating')
plt.xlabel('')
g.set_xticklabels(["Control","Thermedol"])
```

# Patient - Interaction - Beliefs of Effectiveness and doctors' perceived empathy

```python
pat_q = pd.read_csv(os.path.join(fpath,'study1_pat_interaction_belief_data.csv'))
```

```python
%%R -i pat_q
pat_q$Time <- factor(pat_q$Time)
pat_q$Condition <- factor (pat_q$Condition_new)
effectivness_model <- lmer(effectiveness ~ Time*Condition + (1 + Time + Condition | DyadID), data = pat_q)
summary(effectivness_model)
```

```python
%%R
lsmeans(effectivness_model,pairwise~Time*Condition)
```

## Extended Data Figure 2(D)

```python
sns.set(style="ticks", font_scale=1.3)
g = sns.catplot(x="Time_new", y="doc_empathy",hue='Condition',
                   palette={'Thermedol':'dodgerblue','Vaseline':'red'},
                   kind="bar",n_boot=5000, ci=68, aspect=1.3, data=pat_q, legend_out=False)

g = sns.stripplot(x="Time_new", y="doc_empathy", hue='Condition', palette={'Thermedol':'lightskyblue','Vaseline':'lightsalmon'},
                   data=pat_q, alpha=.5, dodge=True)


g.set(ylim=(0, 101))
plt.ylabel("Doctor's perceived empathy")
plt.xlabel('Time')

g.legend(["", ""])
```

```python
%%R -i pat_q
pat_q$Time <- factor(pat_q$Time)
pat_q$Condition <- factor (pat_q$Condition_new)
empathy_model <- lmer(doc_empathy ~ Time*Condition + (1 + Time + Condition | DyadID), data = pat_q)
summary(empathy_model)
```
