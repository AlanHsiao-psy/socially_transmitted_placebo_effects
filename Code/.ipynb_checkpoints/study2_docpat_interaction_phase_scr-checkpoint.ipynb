```python
%matplotlib inline
%config InlineBackend.figure_format = 'retina'
import pandas as pd
import os, glob
import matplotlib.pylab as plt
import numpy as np
import seaborn as sns
import nltools.stats as nlstats
from nltools.stats import downsample
import scipy.signal as signal
from scipy.signal import butter, lfilter, freqz, filtfilt
from scipy.spatial import distance
from scipy import stats
from natsort import natsorted, ns
from pymer4.models import Lmer
from sklearn import metrics
from sklearn.metrics import roc_auc_score
%load_ext rpy2.ipython

# https://stackoverflow.com/questions/25191620/
#   creating-lowpass-filter-in-scipy-understanding-methods-and-units
def butter_bandpass(lowcut, highcut, fs, order=5):
    nyq = 0.5 * fs
    low = lowcut / nyq
    high = highcut / nyq
    b, a = butter(order, [low, high], btype='band')
    return b, a

def butter_bandpass_filter(data, lowcut, highcut, fs, order=5):
    b, a = butter_bandpass(lowcut, highcut, fs, order=order)
    y = filtfilt(b, a, data) # Use filtfilt to avoid phase shift 
    return y

fpath = 'where/you/have/the/data'
```

# Patient - Interaction - SCR

```python
dt = pd.read_csv(os.path.join(fpath,'study2_pat_interaction_scr_auc_data.csv'))
```

```python
patgsr_auc_mean = dt.groupby(['DyadID','Trials','Condition','Counterbalance','Current_Color',
        'Experimenter','Condition_new','Temperature','Site'])['auc'].mean().reset_index()
```

```python
patgsr_auc_mean_forplot = dt.groupby(['DyadID','Condition','Counterbalance'])['auc'].mean().reset_index()
```

```python
patgsr_auc_mean['trials_c'] = patgsr_auc_mean.Trials - patgsr_auc_mean.Trials.mean()
```

```python
%%R -i patgsr_auc_mean
patgsr_auc_mean$Site <- factor(patgsr_auc_mean$Site)
patgsr_auc_mean$Color <- factor(patgsr_auc_mean$Current_Color)
patgsr_auc_mean$Counterbalance <- factor(patgsr_auc_mean$Counterbalance)
patgsr_auc_mean$Condition <- factor (patgsr_auc_mean$Condition_new)
patgsr_auc_mean$Temperature <- factor (patgsr_auc_mean$Temperature)
patgsr_auc_mean$Experimenter <- factor (patgsr_auc_mean$Experimenter)
patgsr_auc_model <- lmer(auc ~ Temperature + Experimenter + Counterbalance*Condition
                         + Site + trials_c + Color + (1 + trials_c + Condition | DyadID)
                         , data = patgsr_auc_mean)
anova(patgsr_auc_model)
```

```python
%%R
lsmeans(patgsr_auc_model,pairwise~Counterbalance*Condition)
```
