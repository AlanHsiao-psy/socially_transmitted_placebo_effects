# Patient - Interaction - SCR AUC

```python
patgsr_auc_dat = pd.read_csv(os.path.join(fpath,'study1_pat_interaction_scr_auc_data.csv'))
```

```python
patgsr_auc_mean = patgsr_auc_dat.groupby(['DyadID','Condition','block'])['auc'].mean().reset_index()
```

```python
%%R -i patgsr_auc_dat
patgsr_auc_dat$Site <- factor(patgsr_auc_dat$Site)
patgsr_auc_dat$Color <- factor(patgsr_auc_dat$Color)
patgsr_auc_dat$Condition <- factor (patgsr_auc_dat$Condition_new)
patgsr_auc_model <- lmer(auc ~ Site + trials_c + Color + Condition + (1 + trials_c + Condition | DyadID)
                         , data = patgsr_auc_dat)
summary(patgsr_auc_model)
```

## Figure 2(A, right panel)

```python
sns.set(style="ticks", font_scale=1.3)
g = sns.catplot(x="Condition", y="auc", palette={'Thermedol':'dodgerblue','Vaseline':'red'},units='DyadID',
                   n_boot=5000, ci=68, data=patgsr_auc_mean, kind="bar", 
                   order=['Vaseline','Thermedol'], aspect=.8)

g = sns.stripplot(x="Condition", y="auc", palette={'Thermedol':'lightskyblue','Vaseline':'lightsalmon'},
                   data=patgsr_auc_mean, order=['Vaseline','Thermedol'],alpha=.5)

g.set(ylim=(-2, 10))
plt.ylabel('AUC (a.u.)')
plt.xlabel('')
g.set_xticklabels(["Control","Thermedol"])
```

# Patient - Interaction - facial expression

```python
pat_ridge_facedata = pd.read_csv(os.path.join(fpath,'study1_pat_interaction_facial_data.csv'))
```

```python
pat_ridge_mean = pat_ridge_facedata.groupby(['DyadID','Condition'])['pred_PainRating'].mean().reset_index()
```

```python
pat_ridge_facedata['trials_c'] = pat_ridge_facedata.N - pat_ridge_facedata.N.mean()
```

```python
%%R -i pat_ridge_facedata
pat_ridge_facedata$Site <- factor(pat_ridge_facedata$Site)
pat_ridge_facedata$Condition <- factor(pat_ridge_facedata$Condition_new)
pat_ridge_facedata$Color <- factor(pat_ridge_facedata$Current_Color)
pat_face_ridge_regression_model <- lmer(pred_PainRating ~ Site + trials_c +  Condition + Color 
                               + (1 + trials_c + Condition |DyadID)
                               , data = pat_ridge_facedata
                            ,control=lmerControl(optimizer="Nelder_Mead",optCtrl=list(maxfun=40000)))
summary(pat_face_ridge_regression_model)
```

## Figure 2(B-2)

```python
sns.set(style="ticks", font_scale=1.3)
g = sns.catplot(x="Condition", y="pred_PainRating", palette={'Thermedol':'dodgerblue','Vaseline':'red'},units='DyadID',
                   n_boot=5000, ci=68, data=pat_ridge_mean, kind="bar", 
                   order=['Vaseline','Thermedol'], aspect=.8)

g = sns.stripplot(x="Condition", y="pred_PainRating", palette={'Thermedol':'lightskyblue','Vaseline':'lightsalmon'},
                   data=pat_ridge_mean, order=['Vaseline','Thermedol'],alpha=.5)

# g.set(ylim=(-0.5, 0.5))
plt.ylabel('Predicted Pain Z-score')
plt.xlabel('')
g.set_xticklabels(["Control","Thermedol"])
```
